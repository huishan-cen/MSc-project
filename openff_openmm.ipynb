## run MD 

```python
from rdkit import Chem
from simtk.openmm.app import PDBFile
from openff.toolkit.topology import Molecule
from openff.toolkit.topology import Topology
from openff.toolkit.typing.engines.smirnoff import ForceField
from simtk import openmm, unit
```

```python
selected_qm9 = {'methanol':'CO', 'ethanal':'CC=O', 'dimethyl_ether':'COC', 'propan-2-ol':'CC(C)O',
                'propan-1-ol':'CCCO', 'methoxyethane':'CCOC', '3,3-dimethylbutan-1-ol':'CC(C)(C)CCO',
                'oxane':'C1CCOCC1','1,4-dioxane':'C1COCCO1', 'hex-3-yne':'CCC#CCC'}
```

```python
for name in selected_qm9.keys():
    smiles = selected_qm9[name]
    # create a PDB file with rdkit
    mol = Chem.MolFromSmiles(smiles)
    mol = Chem.rdmolops.AddHs(mol)
    Chem.AllChem.EmbedMolecule(mol)
    rdkit.Chem.MolToPDBFile(mol, f'Dataset/qm9/{name}.pdb')
    # read the PDB file in openmm
    pdbfile = PDBFile(f'Dataset/qm9/{name}.pdb')
    # use the open force-field to parameterise the molecule
    m = Molecule.from_smiles(smiles)
    ff = ForceField('/usr/local/lib/python3.7/site-packages/openforcefields/offxml/openff-1.3.0.offxml') 
    omm_topology = pdbfile.topology
    off_topology = Topology.from_openmm(omm_topology, unique_molecules=[m])
    system = ff.create_openmm_system(off_topology)
    # MD with openmm
    # Propagate the System with Langevin dynamics.
    temperature = 300*unit.kelvin  # simulation temperature
    integrator = openmm.LangevinIntegrator(temperature, 1/unit.picosecond, 0.001*unit.picosecond)
    platform = openmm.Platform.getPlatformByName('CUDA')
    properties = {'Precision': 'double'}
    # Set up an OpenMM simulation.
    simulation = openmm.app.Simulation(omm_topology, system, integrator, platform, properties)

    # Set the initial positions.
    simulation.context.setPositions(pdbfile.getPositions())

    # Randomize the velocities from a Boltzmann distribution at a given temperature.
    simulation.context.setVelocitiesToTemperature(temperature)

    # Configure the information in the output files.
    simulation.reporters.append(openmm.app.DCDReporter(f'Dataset/qm9/{name}_traj.dcd', 1000))
    simulation.reporters.append(openmm.app.StateDataReporter(f'Dataset/qm9/{name}_log.csv', 1000, 
                                                         time=True, volume=True,temperature=True,kineticEnergy=True, 
                                                         potentialEnergy=True,totalEnergy=True,density=True,speed=True))
    simulation.step(5000000) # a 5 ns equlibration, 5 millions steps
    simulation.step(5000000) # a 5 ns production run, 5 millions steps
    del simulation
```
